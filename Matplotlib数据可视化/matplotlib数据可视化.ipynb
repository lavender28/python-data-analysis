**添加图例 文本 网格**

显示数据的简短描述，更易理解，在图形中添加文本内容，创建和放置数据图例

```python
plt.style.use('seaborn-white')   # 使用样式

plt.rcParams['font.sans-serif'] = 'SimHei'   # win
plt.rcParams['font.sans-serif'] = 'Hiragino Sans GB'   # mac
plt.rcParams['axes.unicode_minus'] = False   # 设置正常显示符号

plt.figure(figsize=(10, 6)) # 创建画布，并指定大小

plt.title('简单图形绘制', fontsize=16)   # 添加标题名称
plt.xlabel('x label', fontsize=14)   # 添加x轴的名称
plt.ylabel('y label', fontsize=14)   # 添加y轴的名称
# plt.xlim(0, 2)   # 确定x轴范围
plt.ylim(-2, 10)   # 确定y轴范围
plt.xticks([0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2])   # 规定x轴刻度
plt.yticks([-2, 0, 2, 4, 6, 8, 10])   # 规定y轴刻度

plt.plot(data, data, 'b--')
plt.text(1.8, 2, s="$y=x$", fontsize=20)   # 添加文本

plt.plot(data, data**2, 'y-')
plt.text(1.75, 4, s="$y=x^2$", fontsize=20)

plt.plot(data, data**3, 'r-.')
plt.text(1.6, 6, s="$y=x^3$", fontsize=20)

plt.grid(ls=':', alpha=0.7)   # 添加网格
plt.legend(labels=('linear', 'quadratic', 'cubic'), loc=2, ncol=3)  # 绘制图例

# plt.show()   # 展示图片
plt.savefig('./pic/simple.png')   # 保存绘制的图形
```

**创建一个figure，包含两个子图**

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
```

```python
plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), color="blue", marker='o')
plt.plot(t2, f(t2), color="black")

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), color="orange", linestyle='--')
```

**一个figure，包含多个子图**

```python
np.random.seed(19680801)

y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure()
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```

**多子图布局**

```python
# Supplot2grid 方法
fig = plt.figure()
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2, rowspan=2) # topleft
ax2 = plt.subplot2grid((3,3), (0,2), rowspan=3)            # right
ax3 = plt.subplot2grid((3,3), (2,0))                       # bottom left
ax4 = plt.subplot2grid((3,3), (2,1))                       # bottom right
fig.tight_layout()
```

**直方图绘制**

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline
# plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['font.sans-serif'] = 'Hiragino Sans GB'

plt.rcParams['axes.unicode_minus'] = False

data = np.load('./data/国民经济核算季度数据.npz', allow_pickle=True)
```

```python
name = data['columns']
name
```

```python
values = data['values']
values
```

```python
plt.figure(figsize=(8, 7))

plt.xlabel('年份')
plt.ylabel('生产总值（亿元）')
plt.title('2000-2017年季度生产总值散点图')
plt.xticks(range(0, 70, 4), values[range(0, 70, 4), 1], rotation=45)

plt.scatter(values[:, 0], values[:, 2], marker='o')

plt.savefig('./tmp/2000-2017年季度生产总值散点图.png')
plt.show()
```

```python
plt.figure(figsize=(8, 7))

plt.xlabel('年份')
plt.ylabel('生产总值（亿元）')
plt.title('2000-2017年季度生产总值折线图')
plt.xticks(range(0, 70, 4), values[range(0, 70, 4), 1], rotation=45)

plt.plot(values[:, 0], values[:, 2], color='red', ls='--')

plt.savefig('./tmp/2000-2017年季度生产总值折线图.png')
plt.show()
```

```python
plt.figure(figsize=(6, 6))

label = ['第一产业', '第二产业', '第三产业']
explode = [0.01, 0.01, 0.2]

plt.pie(values[-1, 3:6], explode=explode, labels=label, autopct='%1.1f%%')

plt.title('2017年第一季度各产业国民生产总值饼图')
plt.savefig('./tmp/2017年第一季度各产业生产总值饼图.png')
plt.show()
```
