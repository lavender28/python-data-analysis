```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
data = np.linspace(0, 2, 100)
```

```python
# 1. 添加画布
plt.figure(figsize=(12, 8))

# 2. 添加内容
plt.title('simple plot', fontsize=18)
plt.xlabel('x label', fontsize=18)
plt.ylabel('y label', fontsize=18)
# plt.xlim(-1, 3)
# plt.ylim(-2, 12)
# plt.xticks([-1, 0, 0.5, 1, 1.5, 2])
# plt.yticks([-2, 0, 2, 4, 6, 8, 10])


plt.plot(data, data, linestyle="--", label='linear')
plt.text(1.8, 2, s='$y=x$', fontsize=18)

plt.plot(data, data**2, linestyle='-.', label='quadratic')
plt.text(1.7, 3.5, s='$y=x^2$', fontsize=18)

plt.plot(data, data**3, linestyle=':', label='cubic')
plt.text(1.6, 6, s='$y=x^3$', fontsize=18)

plt.grid(alpha=0.5)
plt.legend(ncol=3)

# 3. 展示图形
plt.show()
plt.savefig('./pic/sim.png')
```

```python
plt.style.use('seaborn-white')   # 使用样式

plt.rcParams['font.sans-serif'] = 'SimHei'   # win
plt.rcParams['font.sans-serif'] = 'Hiragino Sans GB'   # mac
plt.rcParams['axes.unicode_minus'] = False   # 设置正常显示符号

plt.figure(figsize=(10, 6)) # 创建画布，并指定大小

plt.title('简单图形绘制', fontsize=16)   # 添加标题名称
plt.xlabel('x label', fontsize=14)   # 添加x轴的名称
plt.ylabel('y label', fontsize=14)   # 添加y轴的名称
# plt.xlim(0, 2)   # 确定x轴范围
plt.ylim(-2, 10)   # 确定y轴范围
plt.xticks([0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2])   # 规定x轴刻度
plt.yticks([-2, 0, 2, 4, 6, 8, 10])   # 规定y轴刻度

plt.plot(data, data, 'b--')
plt.text(1.8, 2, s="$y=x$", fontsize=20)   # 添加文本

plt.plot(data, data**2, 'y-')
plt.text(1.75, 4, s="$y=x^2$", fontsize=20)

plt.plot(data, data**3, 'r-.')
plt.text(1.6, 6, s="$y=x^3$", fontsize=20)

plt.grid(ls=':', alpha=0.7)   # 添加网格
plt.legend(labels=('linear', 'quadratic', 'cubic'), loc=2, ncol=3)  # 绘制图例

# plt.show()   # 展示图片
plt.savefig('./pic/simple.png')   # 保存绘制的图形
```
