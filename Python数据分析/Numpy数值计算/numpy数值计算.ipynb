```python
import time
import numpy as np

def python_sum(n):
	a = [i**2 for i in range(n)]
	b = [i**3 for i in range(n)]
	list_sum = []
	for i in range(n):
		list_sum.append(a[i] + b[i])
	return list_sum

def numpy_sum(n):
	a = np.arange(n)**2
	b = np.arange(n)**3
	return a + b
```

```python
%timeit python_sum(1000000)
%timeit numpy_sum(1000000)
```
