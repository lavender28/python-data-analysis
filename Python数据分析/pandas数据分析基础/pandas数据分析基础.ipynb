```python
import pandas as pd

df = pd.read_excel('./data/TOP250.xlsx')
dup_count = df.duplicated().sum()
print(f'重复行数量（保留第一条）: {dup_count}')

# (2) 展示所有重复行（keep=False：把第1次也标记出来）
all_dups = df[df.duplicated(keep=False)]
print(f'涉及重复的行数（含第一次）: {len(all_dups)}')
print(all_dups[['片名', '上映年份', '评分']].to_string())

# (3) 哪部电影重复最多
print('重复最多的电影：')
print(all_dups['片名'].value_counts())
```

```python
# 仅按 片名+上映年份 判断
dup_subset = df.duplicated(subset=['片名', '上映年份']).sum()
print(f'按「片名+上映年份」判断的重复行数: {dup_subset}')
print(f'按所有列判断的重复行数: {df.duplicated().sum()}')

# 展示仅按子集时的重复行
print(df[df.duplicated(subset=['片名', '上映年份'], keep=False)][['片名','上映年份','评分']].to_string())
```
