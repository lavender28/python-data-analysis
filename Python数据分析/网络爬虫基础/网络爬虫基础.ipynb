#### 蹲守robots.txt协议

每个网站根目录下都有 `robots.txt` 文件，明确声明哪些页面允许爬取，哪些禁止。

```python
# 用代码检查 robots.txt
import urllib.robotparser

rp = urllib.robotparser.RobotFileParser()
rp.set_url("https://www.baidu.com/robots.txt")
rp.read()

# 检查某个 URL 是否允许爬取
can_fetch = rp.can_fetch("*", "https://www.baidu.com/news")
print(f"是否允许爬取 /news: {can_fetch}")
```

异常处理与超时设置

```python
import requests
from requests.exceptions import RequestException, HTTPError, ConnectionError, Timeout

def fetch_url(url, headers=None, reties=3):
    """封装一个安全的请求函数，自动重试和错误处理"""
    for attempt in range(1, reties + 1):
        try:
            response = requests.get(url, headers=headers, timeout=10)
            response.raise_for_status()  # 检查 HTTP 错误
            return response
        except HTTPError as http_err:
            print(f"HTTP 错误: {http_err} (尝试 {attempt}/{reties})")
        except ConnectionError as conn_err:
            print(f"连接错误: {conn_err} (尝试 {attempt}/{reties})")
        except Timeout as timeout_err:
            print(f"请求超时: {timeout_err} (尝试 {attempt}/{reties})")
        except RequestException as req_err:
            print(f"请求异常: {req_err} (尝试 {attempt}/{reties})")
    print("所有重试均失败，无法获取数据。")
    return None

# url = "https://www.example.com/books"
url = "https://www.baidu.com/"
headers = {
    "User-Agent": "Mozilla/5.0"
}
response = fetch_url(url, headers=headers)
if response:
    print(f"请求成功，内容长度: {len(response.content)}")
else:
    print("请求失败，跳过该URL。")
```

```python
html_str = "<html><head><title>测试页面</title></head><body><h1>欢迎来到测试页面</h1></body></html>"
print(f"HTML字符串: {html_str}")
```

```python
HTML = """
<html><body>
  <div class="book-list">
    <div class="book-item" id="book-1">
      <h2><a href="/book/1" class="book-link">Python编程：从入门到实践</a></h2>
      <span class="rating">评分：<em class="score">9.1</em></span>
      <span class="price">定价：¥89.00</span>
      <p class="tags"><span class="tag">Python</span><span class="tag">入门</span></p>
    </div>
    <div class="book-item" id="book-2">
      <h2><a href="/book/2" class="book-link">算法导论</a></h2>
      <span class="rating">评分：<em class="score">9.4</em></span>
      <span class="price">定价：¥128.00</span>
      <p class="tags"><span class="tag">算法</span></p>
    </div>
    <div class="book-item" id="book-3">
      <h2><a href="/book/3" class="book-link">机器学习实战</a></h2>
      <span class="rating">评分：<em class="score">8.2</em></span>
      <span class="price">定价：¥69.00</span>
      <p class="tags"><span class="tag">机器学习</span></p>
    </div>
  </div>
</body></html>
"""

tree = etree.HTML(HTML)

# ── 路径选择 ──
# //：匹配文档中任意位置的节点（最常用）
print(tree.xpath("//h2/a/text()"))

# /：从根节点开始（绝对路径）
print(tree.xpath("/html/body/div/div/h2/a/text()"))

# .：当前节点（在循环中用）
for item in tree.xpath("//div[@class='book-item']"):
    title = item.xpath(".//h2/a/text()")[0]
    print(f"书籍标题: {title}")
```

```python
# [@属性='值']：按属性精确筛选
item1 = tree.xpath("//div[@id='book-1']//a/text()")[0]
print(item1)

# [contains(@属性, '部分值')]：包含匹配（更灵活）
# 筛选 class 包含 "book" 的 div
books = tree.xpath("//div[contains(@class, 'book')]//@id")
print(books)

# [位置]：按位置选取（从 1 开始！不是 0）
first = tree.xpath('//div[@class="book-item"][1]//a/text()')[0]
print(first)
last_item = tree.xpath('//div[@class="book-item"][last()]//a/text()')[0]
print(last_item)

# [数值比较]：按文本内容筛选（评分大于 9.0 的书）
high_score = tree.xpath('//em[@class="score"][.>9.0]/text()')
print(high_score)
```

```python
from lxml import etree

# 模拟从 requests 获取到的 HTML 页面
html_content = """
<!DOCTYPE html>
<html>
<head><title>编程书单</title></head>
<body>
  <h1>2024 年度推荐编程书单</h1>
  <div class="book-list">
    <div class="book-item" id="book-1">
      <h2><a href="/book/1" class="book-link">Python编程：从入门到实践</a></h2>
      <p class="author">作者：Eric Matthes</p>
      <p class="publisher">出版社：人民邮电出版社</p>
      <span class="rating">评分：<em class="score">9.1</em></span>
      <span class="price">定价：¥89.00</span>
      <p class="tags"><span class="tag">Python</span><span class="tag">入门</span></p>
    </div>
    <div class="book-item" id="book-2">
      <h2><a href="/book/2" class="book-link">算法导论</a></h2>
      <p class="author">作者：Thomas H. Cormen</p>
      <p class="publisher">出版社：机械工业出版社</p>
      <span class="rating">评分：<em class="score">9.4</em></span>
      <span class="price">定价：¥128.00</span>
      <p class="tags"><span class="tag">算法</span><span class="tag">数据结构</span></p>
    </div>
    <div class="book-item" id="book-3">
      <h2><a href="/book/3" class="book-link">机器学习实战</a></h2>
      <p class="author">作者：Peter Harrington</p>
      <p class="publisher">出版社：人民邮电出版社</p>
      <span class="rating">评分：<em class="score">8.2</em></span>
      <span class="price">定价：¥69.00</span>
      <p class="tags"><span class="tag">机器学习</span><span class="tag">算法</span></p>
    </div>
  </div>
</body>
</html>
"""

# ── 构建解析树 ──
tree = etree.HTML(html_content)

# ── 结构化提取所有书籍信息 ──
books = []
for item in tree.xpath('//div[@class="book-item"]'):
    book = {
        'id':        item.get('id'),
        'name':      item.xpath('.//a[@class="book-link"]/text()')[0].strip(),
        'link':      item.xpath('.//a[@class="book-link"]/@href')[0],
        'author':    item.xpath('.//p[@class="author"]/text()')[0].replace('作者：', '').strip(),
        'publisher': item.xpath('.//p[@class="publisher"]/text()')[0].replace('出版社：', '').strip(),
        'score':     float(item.xpath('.//em[@class="score"]/text()')[0]),
        'price':     item.xpath('.//span[@class="price"]/text()')[0].replace('定价：', '').strip(),
        'tags':      item.xpath('.//span[@class="tag"]/text()'),
    }
    books.append(book)

# ── 输出结果 ──
print(f"共提取到 {len(books)} 本书籍：\n")
for b in books:
    print(f"  [{b['id']}] 《{b['name']}》")
    print(f"       作者: {b['author']}  |  出版社: {b['publisher']}")
    print(f"       评分: {b['score']}   |  价格: {b['price']}")
    print(f"       标签: {', '.join(b['tags'])}\n")

# ── 数据筛选（评分 >= 9.0 的书）──
high_rated = [b for b in books if b['score'] >= 9.0]
print(f"评分 ≥ 9.0 的书籍：{[b['name'] for b in high_rated]}")
print(f"评分最高: 《{max(books, key=lambda x: x['score'])['name']}》")
```

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
  <head><title>编程书单</title></head>
  <body>
    <h1 class="main-title">2024年推荐书单</h1>
    <div class="book-list">
      <div class="book-item" id="book-1">
        <h2><a href="/book/1">Python编程：从入门到实践</a></h2>
        <p class="author">Eric Matthes</p>
        <em class="score">9.1</em>
      </div>
      <div class="book-item" id="book-2">
        <h2><a href="/book/2">算法导论</a></h2>
        <p class="author">Thomas H. Cormen</p>
        <em class="score">9.4</em>
      </div>
    </div>
  </body>
</html>
"""

soup = BeautifulSoup(html_doc, 'lxml')

# ── 直接用标签名访问（返回第一个匹配）──
print(soup.title)  # <title>编程书单</title>
print(soup.title.name)  # title
print(soup.title.text)  # 编程书单
print(soup.title.string)  # 编程书单 仅限于标签内只有文本时使用

print(soup.h1)  # <h1 class="main-title">2024年推荐书单</h1>
print(soup.h1['class'])  # ['main-title'] 返回属性值列表
print(soup.h1.get('class'))  # ['main-title'] 推荐使用 get 方法

# ── 树形导航 ──
div = soup.find('div', class_='book-list')  # <div class="book-list">...</div>
print("父节点:", div.parent.name)  # body
print("第一个子节点:", div.div.h2.a.text) 

# 遍历所有直接子节点
print("book-list的直接子节点：")
for child in soup.find('div', class_='book-list').children:
    if child.name:  # 过滤掉文本节点
        print(f"  {child.name} - id: {child.get('id')}")

# 获取下一个/上一个兄弟节点
first_book = soup.find('div', class_='book-item')
print("第一个书籍标题:", first_book.h2.a.text)  # Python编程：从入门到实践
print("下一个兄弟节点标题:", first_book.find_next_sibling('div').h2.a.text)  # 算法导论
```

find 和 find_all 查找

```python
# ── find()：查找第一个匹配 ──
print("第一个<p>标签:", soup.find('p'))                          # 第一个 <p>
print("第一个class为author的p标签:", soup.find('p', class_='author'))         # class=author 的第一个 <p>
print("id为book-1的div:", soup.find('div', id='book-1'))           # id=book-1 的 div
print("class为score的em:", soup.find('em', class_='score'))         # class=score 的 em
```

```python
# ── 提取文本和属性 ──
for item in soup.find_all('div', class_='book-item'):
    name   = item.find('a').text
    link   = item.find('a')['href']
    author = item.find('p', class_='author').text
    score  = item.find('em', class_='score').text
    print(f"{name} | {author} | {score}")
```

CSS 选择器（select）

```python
# ── select()：返回所有匹配（列表）──
# ── select_one()：返回第一个匹配 ──

# 标签选择器
soup.select('h2')                     # 所有 h2

# 类选择器
soup.select('.book-item')             # class=book-item 的元素
soup.select('.book-item .author')     # book-item 内的 author 类元素

# ID 选择器
soup.select('#book-1')                # id=book-1 的元素

# 属性选择器
soup.select('[href]')                 # 有 href 属性的元素
soup.select('a[href="/book/1"]')      # href="/book/1" 的 a
soup.select('[id^="book"]')           # id 以 "book" 开头的元素

# 子元素选择器（>：直接子元素）
soup.select('div.book-list > div')    # book-list 的直接 div 子元素

# 后代选择器（空格：任意层级后代）
soup.select('div.book-item a')        # book-item 内所有 a

# 组合选择器（逗号）
soup.select('h1, h2')                 # h1 和 h2

# ── 实战：提取所有书籍数据 ──
print("─── CSS 选择器提取书籍信息 ───")
for item in soup.select('div.book-item'):
    name   = item.select_one('h2 a').text
    link   = item.select_one('h2 a')['href']
    author = item.select_one('p.author').text
    score  = float(item.select_one('em.score').text)
    print(f"《{name}》  {author}  {score}分  {link}")
```
