**标量化和向量化运算**

```python
def pySum(n):
    a = [i**2 for i in range(n)]
    b = [i**3 for i in range(n)]
    c = []
    for i in range(n):
        c.append(a[i] + b[i])
    return c
```

```python
%timeit pySum(10000)
```

```python
import numpy as np
def npSum(n):
    a = np.arange(n)**2
    b = np.arange(n)**3
    c = a + b
    return c
```

```python
%timeit npSum(10000)
```
