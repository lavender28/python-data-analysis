```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('./data/meal_order_detail.xlsx', sheet_name=0)
```

```python
data.shape
```

```python
data.head()
```

```python
data['date'] = [i.date() for i in pd.to_datetime(data['place_order_time'])]
data[['date', 'counts', 'amounts']].head()
```

```python
data_grouped = data[['date', 'counts', 'amounts']].groupby('date')
```

```python
data_grouped.size()
```

```python
day_mean = data_grouped.agg({'amounts': 'mean'})
```

```python
day_mean
```

```python
day_median = data_grouped.agg({'amounts': 'median'})
```

```python
day_median
```
