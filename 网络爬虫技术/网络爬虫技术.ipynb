# requests库的基本使用

```python
# 导入模块
import requests

# 指定url
url = 'http://www.wzbc.edu.cn/'

# 发起请求
response = requests.get(url=url)

# 获取响应数据
page_text = response.text

# 持久化存储
with open('./wzbc.html', 'w', encoding='utf-8') as fp:
    fp.write(page_text)
```

```python
from lxml import etree

html_str = '''
<div>
  <ul>
    <li class="item1"><a href="lang1">Python</a></li>
    <li class="item2"><a href="lang2">Java</a></li>
    <li class="item3"><a href="lang3">Go</a>
    <li class="item4"><a href="lang4">JavaScript</a></li>
    <li class="item5"><a href="lang5">Ruby</a></li>
  </ul>
</div>
'''
# 创建解析对象
html = etree.HTML(html_str)
# 编写xpath表达式
xpath_exp = '//a/text()'
# 提取文本数据，以列表形式输出
r_list = html.xpath(xpath_exp)

print(r_list)
```

```python
import requests
from lxml import etree
import json

class DouBanTop250():
    def __init__(self):
        self.url = 'https://movie.douban.com/top250'
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 \
            (KHTML, like Gecko) Chrome/98.0.4758.102 Mobile Safari/537.36'
        }
    def getHtml(self):
        try:
            r = requests.get(url=self.url, headers=self.headers)
            r.raise_for_status()
            return r.text
        except:
            print('数据获取失败')
    def parseHtml(self):
        parse_html = etree.HTML(self.getHtml())
        movie_info = parse_html.xpath('//div[@class="info"]')
        movie_list = []
        for info in movie_info:
            movie_item = {}
            movie_item['name'] = info.xpath('.//div[@class="hd"]//span[@class="title"]/text()')[0].strip()
            movie_item['star'] = info.xpath('.//div[@class="bd"]/p/text()')[0].split(':')[-1].strip()
            movie_item['rating_num'] = info.xpath('.//span[@class="rating_num"]/text()')[0].strip()
            print(movie_item)
            movie_list.append(movie_item)
        return movie_list
    def saveHtml(self):
        data = self.parseHtml()
        with open('movies.json', 'w', encoding='utf8') as fp:
            json.dump(data, fp, ensure_ascii=False, indent=4)
if __name__ == '__main__':
    spider = DouBanTop250()
    spider.saveHtml()
```

```python
import requests
from bs4 import BeautifulSoup
import json

class BookTop250():
    def __init__(self):
        self.url = 'https://book.douban.com/top250?start=25'
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 \
            (KHTML, like Gecko) Chrome/98.0.4758.102 Mobile Safari/537.36'
        }
    def getHtml(self):
        try:
            r = requests.get(url=self.url, headers=self.headers)
            r.raise_for_status()
            return r.text
        except:
            print('数据获取失败')
    def parseHtml(self):
        soup = BeautifulSoup(self.getHtml())
        comment_list = []
        comment_item_list = soup.find_all('tr')
        for comment_item in comment_item_list:
            name = comment_item.find('div', class_='pl2').a['title']
            author = comment_item.find('p', class_='pl').string.split('/')[0].strip()
            rating_num = comment_item.find('span', class_='rating_nums').string.strip()
            comment = comment_item.find('span', class_='inq').string.strip()
            comment_list.append({
                'name':name,
                'author':author,
                'rating_num':rating_num,
                'comment':comment})
        return comment_list
        
    def saveHtml(self):
        data = self.parseHtml()
        with open('books.json', 'w', encoding='utf8') as fp:
            json.dump(data, fp, ensure_ascii=False, indent=4)
if __name__ == '__main__':
    spider = BookTop250()
    spider.saveHtml()
```
